Bouncing off our bdsf notebook, let's try Aegean on the same snip. Aegean doesn't have polarised source-finding, so we'll compute a per-channel P image and source-find on that instead.

```python
import AegeanTools.source_finder as aegean
from AegeanTools.models import GlobalFittingData
from AegeanTools.catalogs import save_catalog
import astropy.io.fits
import logging
import pandas

logger = logging.root
```

```python
sf = aegean.SourceFinder(global_data=GlobalFittingData(), log=logger)
```

```python
sf.load_globals(
    '/home/alger/possum/racs/squished/P_04.fits')
sf.find_sources_in_image(
    '/home/alger/possum/racs/squished/P_04.fits', cores=1)
```

```python
save_catalog('/home/alger/possum/alger/racs-4-aegean-p.csv', sf.sources)
```
